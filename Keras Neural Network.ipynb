```python
from __future__ import absolute_import, division, print_function

import tensorflow as tf
import sklearn.model_selection
from tensorflow import keras
from tensorflow.keras import layers

import numpy as np
import pandas as pd
```

```python
dataset_path = 'Admission_Predict_Ver1.1.csv'
raw_dataset = pd.read_csv(dataset_path, sep=",")

raw_dataset.info()
raw_dataset.head()
```

---
# Regresi√≥n

```python
dataset = raw_dataset.copy()

x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values.reshape(-1, 1)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, shuffle=True, random_state=0)
```

```python
x
```

```python
y
```

```python
def build_model():   
    model = keras.models.Sequential()
    model.add(keras.layers.Dense(32, activation = 'relu', input_shape = (x_train.shape[1],)))
    model.add(keras.layers.Dense(1, activation = 'sigmoid'))
    model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['acc'])
    
    return model
```

```python
model = build_model()
model.summary()
```

```python
history = model.fit(x_train, y_train, 
                    epochs=40,
                    validation_data=(x_test, y_test))
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
# Error cuadratico medio
from sklearn.metrics import mean_squared_error
cm = mean_squared_error(y_test, predictions)
cm
```

# Clasificaci√≥n

```python
# Hacer una red con 5 salidas y ver cual se prende???
```

```python
def classify_predictions(predictions):
    arr = []
    for pred in predictions:
        pred = pred[0]
        if 0.9 < pred < 1:
            pred_class = "Very Good"
        if 0.8 < pred < 0.9:
            pred_class = "Good"
        if 0.65 < pred < 0.8:
            pred_class = "Medium"
        if 0.5 < pred < 0.65:
            pred_class = "Bad"
        if 0 < pred < 0.5:
            pred_class = "Very Bad"

        arr.append([pred_class])


    return np.array(arr)

classified_predictions = classify_predictions(predictions)
classified_y_test = classify_predictions(y_test)
classified_predictions, classified_y_test
```

```python
# % de casos de √©xito 
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(classified_y_test, classified_predictions)
print('Evaluated accuracy: ', (accuracy * 100))
```
